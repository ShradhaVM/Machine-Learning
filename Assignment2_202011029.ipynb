**Obtaining Sample means and Sample Variance**

```
# Calculating Normal Distributions for the parameters given
nd1 = np.random.normal(10,np.sqrt(20),200)
nd2 = np.random.normal(20,np.sqrt(25),200)

# Calculating Sample means and Sample variance from above distributions
s_mean1 = np.mean(nd1)
s_var1 = np.var(nd1)
s_mean2 = np.mean(nd2)
s_var2 = np.var(nd2)
```

```
import numpy as np
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split 
from scipy.stats import multivariate_normal
```

**Loading iris data and dividing into test classes**

```
iris = load_iris() 
data_list = iris.data 
target_list = iris.target 

orig_set1 = [*target_list[0:10]]
orig_set2 = [*target_list[50:60]]
orig_set3 = [*target_list[100:110]]
print("The Ideal classification :", orig_set1 + orig_set2 + orig_set3)
```
