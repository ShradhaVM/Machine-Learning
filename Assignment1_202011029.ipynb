```
import numpy as np
import matplotlib.pyplot as plt
from math import log, cos, sin, sqrt, pi
```

Draw 100 samples from uni-variate Normal distribution, plot its histogram. Compute pdf for univariate normal distribution and compare it with histogram. Repeat experiment with 500, 1000 data points.

```
mean, sd = 0, 1

# Normal distributions with different number of samples
norm_dist1 = np.random.normal(mean, sd, 100)
norm_dist2 = np.random.normal(mean, sd, 500)
norm_dist3 = np.random.normal(mean, sd, 1000)

plt.subplot(231)
n1, bins1, cont1 = plt.hist(norm_dist1, 25, density=True)
# PDF for normal distribution with 100 samples
pdf1 = 1/(sd * np.sqrt(2 * np.pi)) * np.exp( - (bins1 - mean)**2 / (2 * sd**2))
plt.plot(bins1, pdf1, linewidth = 2, color = 'r')
plt.title('With 100 samples')

plt.subplot(232)
n2, bins2, cont2 = plt.hist(norm_dist2, 25, density=True)
# PDF for normal distribution with 500 samples
pdf2 = 1/(sd * np.sqrt(2 * np.pi)) * np.exp( - (bins2 - mean)**2 / (2 * sd**2))
plt.plot(bins2, pdf2, linewidth = 2, color = 'r')
plt.title('With 500 samples')

plt.subplot(233)
n3, bins3, cont3 = plt.hist(norm_dist3, 25, density=True)
# PDF for normal distribution with 1000 samples
pdf3 = 1/(sd * np.sqrt(2 * np.pi)) * np.exp( - (bins3 - mean)**2 / (2 * sd**2))
plt.plot(bins3, pdf3, linewidth = 2, color = 'r')
plt.title('With 1000 samples')

plt.show()
```

Draw 100 samples from bi-variate Normal distribution.

```
# Implementing bi-variate normal distribution by defining a function
def hist_bivariate(mean, cov, samples) : 
  x, y = np.random.multivariate_normal(mean, cov, samples).T
  m = np.linspace(-5,5,20)
  n = np.linspace(-5,5,20)
  freq, x_edge , y_edge = np.histogram2d(x, y, bins = (m,n))
  X , Y = np.meshgrid(x_edge[:-1], y_edge[:-1])
  X = X.ravel()
  Y = Y.ravel()

# Plotting the 3D histogram
fig = plt.figure()
ax = fig.add_subplot( projection = '3d')
hist_bivariate([0,0], [[1,0],[0,1]], 100)
ax.bar3d(X, Y, 0, 1, 1, freq.ravel(), zsort = 'average')
plt.show()
```

Implement central limit theorem.

```
# Implementing CLT using randomly generated discrete values
disc_mean = [] # disc_mean will store means for generated random distributions
for i in range(500) : 
    disc_mean.append(np.mean(np.random.randint(0, 50, 12)))

# Plotting the histogram
plt.hist(disc_mean, 20, density = True)
plt.title('CLT with Random Discrete Selection from 12 samples')
plt.show()
```

Draw 12 samples from Uni[0,1] as a sample population and perform CLT.

```
# Implementing CLT using randomly but uniformly generated values
uni_mean = [] # uni_mean will store means for generated random distributions
for i in range(500) : 
    uni_mean.append(np.mean(np.random.uniform(0, 1, 12)))
plt.hist(uni_mean, 20, density = True)
plt.title('CLT with Random Uniform Selection from 12 samples')
plt.show()
```

Use Box-Muller method to achieve Normal distribution.

```
# generating random values for box muller method
x1 = np.random.rand(500)
x2 = np.random.rand(500)

# Box muller method using both cosine and sine functions
c = [sqrt(-2*log(x1[i]))*cos(2*pi*x2[i]) for i in range(500)]
s = [sqrt(-2*log(x1[i]))*sin(2*pi*x2[i]) for i in range(500)]

# Plotting the histogram
plt.subplot(221)
plt.hist(c, 20, density = True)
plt.title('Using cosine function')

plt.subplot(222)
plt.hist(s, 20, density = True)
plt.title('Using sine function')

plt.show()
```
